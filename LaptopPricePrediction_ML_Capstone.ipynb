```python
#outlier treatment

Q1 = lp['Inches'].quantile(0.25)
Q3 = lp['Inches'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier criteria
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Capping the outliers
lp['Inches']=lp['Inches'].clip(lower_bound, upper_bound)
```

```python
#outlier treatment
Q1 = lp['Weight'].quantile(0.25)
Q3 = lp['Weight'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier criteria
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Capping the outliers
lp['Weight']=lp['Weight'].clip(lower_bound, upper_bound)

lp['Weight'].plot(kind='box')
```

```python
#droping empty values in price column
lp.dropna(subset=['Price'],inplace=True)
lp.info()
```

```python
#treating the memory column
#converting from object to str to apply str.replace and then replacing GB,TB and finally finally splitting into 2 columns
lp['Memory'] = lp['Memory'].astype(str).replace(r'\.0', '')
lp["Memory"] = lp["Memory"].str.replace('GB', '')
lp["Memory"] = lp["Memory"].str.replace('TB', '000')
splitted_memory = lp["Memory"].str.split("+", n = 1, expand = True)

lp["first"]= splitted_memory[0]
lp["first"]=lp["first"].str.strip()

lp["second"]= splitted_memory[1]
lp["second"].fillna("0", inplace = True)
```

```python
#creating binary columns for [0]th position HDD SDD Hybrid and Flash storage

lp["1HDD"] = lp["first"].apply(lambda x: 1 if "HDD" in x else 0)
lp["1SSD"] = lp["first"].apply(lambda x: 1 if "SSD" in x else 0)

#converting str to int
lp['first'] = lp['first'].str.replace(r'\D', '', regex=True)
lp['first'] = lp['first'].replace('', '0')
lp["first"] = lp["first"].astype(int)
```

```python
#creating binary columns for [1]th position HDD SDD Hybrid and Flash storage

lp["2HDD"] = lp["second"].apply(lambda x: 1 if "HDD" in x else 0)
lp["2SSD"] = lp["second"].apply(lambda x: 1 if "SSD" in x else 0)

#converting str to int
lp['second'] = lp['second'].str.replace(r'\D', '',regex=True)
lp["second"] = lp["second"].astype(int)
```

```python
#performing multiplication to get a column that is aligned with the OG memory column and storing them in a new columns respectively

lp["HDD"]=(lp["first"]*lp["1HDD"]+lp["second"]*lp["2HDD"])
lp["SSD"]=(lp["first"]*lp["1SSD"]+lp["second"]*lp["2SSD"])
```

```python
lp.drop(columns=["Memory","1HDD","1SSD","2HDD","2SSD","first","second"],inplace=True)
lp
```

```python
# prompt: I want to remove spaces inbetween the Gpu data fields. Like i want it as IntelIrisPlusGraphics640

lp['Gpu']=lp['Gpu'].str.replace(' ', '')
lp.head()
```

```python
# separate independent variable and a dependent variable
X=lp.drop(columns=["Price"])
y=lp["Price"]
```

```python
# split the dataset into train and test

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#testing lin regression model
y_predicted=linear_reg.predict(X_test)
y_predicted
```

```python
#evaluating using rmse
print(mean_squared_error(y_test,y_predicted))
mse=mean_squared_error(y_test,y_predicted)
rmse=np.sqrt(mse)
print(rmse)
```

```python
#testing random forest
y_predicted=rf.predict(X_test)
y_predicted
```

```python
def prediction(Company,	TypeName,	Inches,	Ram,	Gpu,	OpSys,	Weight, screen_res, Cpu_brand,  HDD, SSD):
  inputs_df=pd.DataFrame([[Company,	TypeName,	Inches,	Ram,	Gpu,	OpSys,	Weight, screen_res, Cpu_brand,  HDD, SSD]],columns=["Company",	"TypeName",	"Inches",	"Ram",	"Gpu",	"OpSys",	"Weight", "screen_res", "Cpu_brand",  "HDD", "SSD"])
  # Convert object columns to categorical if they represent categories
  for col in inputs_df.select_dtypes(include=['object']):
    inputs_df[col] = inputs_df[col].astype('category')
  output=xgb.predict(inputs_df) # Enable categorical support
  return output[0]
```

```python
#building an UI using gradio application to get the Price for the input features
import gradio as gr
company_input=gr.Dropdown(label="Company",choices=["Apple","Asus","Dell","Google","Microsoft","Ryzer","Samsung","Xiomi"])
typename_input=gr.Dropdown(label="TypeName",choices=["Ultrabook","Notebook","2 in 1 Convertible","Gaming","Netbook","Workstation"])
inches_input=gr.Number(label="Inches without decimal")
ram_input=gr.Number(label="Ram")
gpu_input=gr.Dropdown(label="Gpu",choices=["Intel Iris Plus Graphics 640","Intel HD Graphics 620","AMD Radeon R5 M330"])
opsys_input=gr.Dropdown(label="Operating System",choices=["macOS","Windows 7","Windows 10","No OS"])
weight_input=gr.Number(label="Weight in KG without decimal")
screenres_input=gr.Dropdown(label="Screen Resolution",choices=["IPS","Full HD","4K","Touchscreen"])
cpubrand_input=gr.Dropdown(label="Cpu Brand",choices=["Intel Core i3","Intel Core i5","Intel Core i7","AMD A6","AMD A9"])
hdd_input=gr.Number(label="HDD x1000")
ssd_input=gr.Number(label="SSD")
```

```python
demo=gr.Interface(fn=prediction,inputs=[company_input,typename_input,inches_input,ram_input,gpu_input,opsys_input,weight_input,screenres_input,cpubrand_input,hdd_input,ssd_input],outputs=gr.Number(label="Predicted Price"))
demo.launch()
```
